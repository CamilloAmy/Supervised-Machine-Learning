```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Generate random data points
np.random.seed(0)
X = 2 * np.random.rand(10, 1)
y = 4 + 3 * X + np.random.randn(10, 1)

# Fit linear regression model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Visualize the data and the linear regression line
plt.figure(figsize = (16,8))
plt.subplot(1,2,1)
plt.scatter(X, y, color='blue', label = 'Trained data')
plt.plot(X, y_pred, color='red', label = 'Predicted model') 
for i in range(len(X)):
    plt.vlines(X[i], y[i], y_pred[i], colors='black', linestyles='dotted')

plt.title('Regression Models')
plt.xlabel('X')
plt.ylabel('y')

##------------------------###

# Fit polynomial regression model
poly_features = PolynomialFeatures(degree=7)  
X_poly = poly_features.fit_transform(X)
model = LinearRegression()
model.fit(X_poly, y)

# Predictions
X_pred_p = np.linspace(min(X), max(X), 10).reshape(-1, 1)  # Generate points for smooth curve
X_pred_poly = poly_features.transform(X_pred_p)
y_pred_p= model.predict(X_pred_poly)
plt.subplot(1,2,2)
plt.scatter(X, y, color='blue', label = 'Trained data')
plt.plot(X, y_pred, color='red', label = 'Predicted model') 
for i in range(len(X)):
    plt.vlines(X[i], y[i], y_pred[i], colors='black', linestyles='dotted')

plt.plot(X_pred_p, y_pred_p, color='green', label='Predicted Model')
plt.legend()


plt.title('Regression Models')
plt.xlabel('X')
plt.ylabel('y')
plt.show()


```
